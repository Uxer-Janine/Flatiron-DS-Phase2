## Summary statistics - Descriptive statistics 
- Conclusions about the sample alone based on the sample. 
- Describes and summarizes data in a meaningful way:

    a) **Measures of central tendency** - mean, mode, median
    
    b) **Measures of dispersion** - variance,standard deviation, range, IQR.
    
    c) **Frequency tables** 
    
    d) **Charts** 
    
    e) **Contingency tables** - Two or more categorical variables. Includes total of both.

#### Correcting the non-sensical value

Replace the question marks with something that python can easily interprete. 

```python
# replace the question marks in the entire data frame 

df = df.replace('?', np.nan)
```

```python
# confirming that our values have been replaced

df.ca.unique() 
```

### Changing the ca column data type 
Since we have established that the values in ca are actually integeres, we now make a point of changing it's data type from "object" to "float". We will get to `thal` in a short while. 

```python
# changing df.ca datatyes to float 

df2.ca = df2.ca.astype("float")
```

```python
# confirming our changes 

df2.info()
```

```python
# df2 summary statistics 

df2.describe(include = "all")
```

**Observation:**

As you can see, we have successfully changed the ca data type to float, and we have even gotten it's summary statistics. 

## Dealing with categorical data 
Some values in the dataframe are categorical but have been recorded as quantitative values instead.

### cp column
Changing cp values from numerals to the different chest pain levels.

```python
# **cp:** chest pains types
#     - value 1: typical angina 
#     - Value 2: atypical angina 
#     - Value 3: non-anginal pain 
#     - value 4: asymptomatic

df2.cp = df2.cp.replace({1.0 : "typical angina", 2.0 : "atypical angina", 3.0 : " non-anginal pain", 4.0 : "asymptomatic"})
```

```python
# confirm that changes have been implemented 

df2.cp.unique()
```

**Observation:** 

The cp values have been changed from numerical values (1,2,3,4) to the different pain levels ('typical angina', 'asymptomatic', ' non-anginal pain',
       'atypical angina')

## Dealing with categorical data/values
Applicable for ordinal data

```python
#  value 1: upsloping
#     - value 2: flat
#     - value 3: downsloping
        
df2.slope.replace({1.0 : "upsloping", 2.0 : "flat", 3.0 :"downsloping"}, inplace = True)

#confirmation 
df2.slope.unique()
```

**Instruct python to treat it as a categorical data** 

```python
df2.slope = pd.Categorical(df2.slope,["upsloping", "flat", "downsloping"], ordered=True)
```

```python
df2.slope.cat.codes # category codes 
```

```python
sns.countplot(x = df2.slope);
```
