# 📊 Phase 2 Statistics Survival Map – Flatiron

Welcome to your personal Stats Map — this is your no-pressure, steady-paced path to understanding the most essential stats concepts in Data Science.

---

## ✅ Goals
- Understand the **“why”** and **“when”** of key statistical tools
- Get comfortable interpreting test results (not just coding them)
- Know when to **ask AI** vs when to **lean in and learn**
- Build confidence in presenting results clearly

---

## 🧱 Core Concepts (One per Day)

### 📌 Descriptive Statistics
- Mean, median, mode
- Range, standard deviation, variance
- When to use what (summary, central tendency, spread)

### 📌 Data Distributions
- Normal vs skewed
- Outliers & Z-scores
- Visuals: histograms, boxplots, density plots

### 📌 Probability Basics
- Independent vs dependent events
- Intro to probability rules (AND, OR)
- Real-life examples (e.g., flipping coins vs drawing cards)

### 📌 Central Limit Theorem (CLT)
- Sample means form a normal distribution (eventually)
- Big deal for hypothesis testing and p-values

### 📌 Confidence Intervals
- The “neighborhood” where we expect the true value
- Interpreting 95% CI in simple language

### 📌 Hypothesis Testing
- Null vs alternative hypothesis
- p-values (how surprised we are)
- Statistical significance (and what it actually means)

### 📌 T-Tests
- One-sample, two-sample, paired
- When to use which (comparing groups)

### 📌 Chi-Squared Test
- Used for **categorical** data
- Checks if distributions differ (e.g., gender vs preference)

### 📌 Correlation vs Causation
- Pearson's r (strength & direction)
- Scatter plots, outliers, and common traps

### 📌 Experimental Design & A/B Testing
- Control vs treatment
- Randomization, sample size
- Measuring success: conversion, lift, impact

### 📌 Cross Validation
- Train-test splits & folds
- Preventing overfitting
- Real-world model validation

### 📌 Bayesian Thinking (Optional)
- Updates beliefs with new evidence
- Used more in advanced/uncertain decision-making
